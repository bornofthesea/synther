```python
fname = "/content/drive/MyDrive/Pesquisa/PhD/Data/tabcnoalpha2025.dat"

aData = pd.read_table(fname, sep='\s+', header=0, skip_blank_lines=True, index_col=False)

fname = "/content/drive/MyDrive/Pesquisa/PhD/Data/newtab.dat"

pskData = pd.read_table(fname, sep='\s+', header=0, skip_blank_lines=True, index_col=False, na_values = "---")

fullData = pd.concat([pskData, aData.drop(columns=['ID2M', 'IDint', 'fe'])], axis=1)

#fullData.to_csv('/content/drive/MyDrive/Pesquisa/PhD/Data/fullData.dat', sep='\t', index=False)
```

```python
fname = "/content/drive/MyDrive/Pesquisa/PhD/Data/brauner.tsv"

braunerData = pd.read_table(fname, sep='\t', comment='#')
display(braunerData)
```

```python
plt.plot(fullData['[Fe/H]'], fullData['S'], color='tab:orange', marker='.', linestyle='None', label = 'Barbuy et al.')
plt.plot(fullData['[Fe/H]'], fullData['Sapo'], color='tab:red', marker='.', linestyle='None',  label = 'APOGEE')
plt.plot(braunerData['[Fe/H]'], braunerData['[S/Fe]'], color='tab:cyan', marker='.', linestyle='None',  label='Brauner')
plt.xlabel('[Fe/H]')
plt.ylabel('[S/Fe]')
plt.legend()
```

```python
plt.plot(fullData['[Fe/H]'], fullData['K'], color='tab:orange', marker='.', linestyle='None', label = 'Barbuy et al.')
plt.plot(fullData['[Fe/H]'], fullData['Kapo'], color='tab:red', marker='.', linestyle='None',  label = 'APOGEE')
plt.xlabel('[Fe/H]')
plt.ylabel('[K/Fe]')
plt.legend()
```

```python
plt.plot(fullData['[Fe/H]'], fullData['P'], color='tab:orange', marker='.', linestyle='None', label = 'Barbuy et al.')
plt.plot(braunerData['[Fe/H]'], braunerData['[P/Fe]'], color='tab:cyan', marker='.', linestyle='None',  label='Brauner')
plt.xlabel('[Fe/H]')
plt.ylabel('[P/Fe]')
plt.legend()
```
