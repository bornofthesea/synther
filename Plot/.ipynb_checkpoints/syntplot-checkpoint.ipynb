```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def setxTicks(ax, N = 5):
    xmin, xmax = ax.get_xlim()
    custom_ticks = np.linspace(xmin, xmax, N, dtype=int)
    ax.set_xticks(custom_ticks)
    ax.set_xticklabels(custom_ticks)

def linePlot(f1, f2, f3, normp1_1, normp1_2, normk1_1, norms1_1, norms1_2):
    """
    This function will plot the region around the lines of P, S and K in the IR with the input synthetic spectra
    and additional spectra (currently only from APOGEE)
    TO BE DONE: USE A LIST OF FILES TO ALLOW FOR MULTIPLE PLOTS
    :param f1: Name of the file containing the synthetic spectra used for the P lines, around 15600 and 16600, as to include the PI line in 16482.9, in flux per wavelength
    :param f2: Name of the file containing the synthetic spectra in the range for the S and K lines, from 15000 to 16000.
    :param f3: Name of the file containing the observed spectra from APOGEE. This spectra is not normalized, so additional normalization parameters should be entered as well
    :param norm*: Normalization parameters to be divided from flux values, one for each range around the interested lines, to be entered manually for each input star.
    """

    # Set the directories in case of program running in a different computer.
    syntdir = "/home/morgan/Turbospectrum2019/COM-v19.1/syntspec/" # Refer to synthetic spectra directory from turbospectrum
    dr14dir = "/home/morgan/PhD/Data/apogee/dr14/" # Observed spectra from APOGEE DR14
    dr16dir = "/home/morgan/PhD/Data/apogee/dr14/" # DR16
    
    fname = syntdir + f1
    adata = pd.read_csv(fname, sep="\s+", header=None, names=["Wavelength", "Flux"])
    
    fname = syntdir + f2
    bdata = pd.read_csv(fname, sep="\s+", header=None, names=["Wavelength", "Flux"])
    
    fname = dr14dir + f3
    apogdata = pd.read_csv(fname, sep="\s+", header=None, names=["Wavelength", "Flux"], comment="#")

    # These set of parameters are the limits of each plot, around the zones containing our lines of interest
    limp1 = (15702,15720)
    limp1_2 = (16478,16494)
    limk1 = (15160, 15170)
    lims1 = (15465, 15473)
    lims1_2 = (15472, 15482)

    # PI lines plots
    fig, ax = plt.subplots(2)
    ax[0].set_title("PI lines")
    ax[0].plot(adata["Wavelength"], adata["Flux"], color="black", label="")
    ax[0].plot(apogdata["Wavelength"], apogdata["Flux"]/normp1_1, color="red")
    ax[0].set_xlim(*limp1)
    ax[0].set_ylim(0.69,1.01)
    ax[0].vlines(x = 15711.6, ymin = 0.75, ymax=0.85, color="purple") # PI 15711.6
    setxTicks(ax[0])
    ax[1].plot(adata["Wavelength"], adata["Flux"], color = "black")
    ax[1].plot(apogdata["Wavelength"], apogdata["Flux"]/normp1_2, color="red")
    ax[1].set_xlim(*limp1_2)
    ax[1].set_ylim(0.69,1.01)
    ax[1].set_xlabel("Wavelength (A)")
    ax[1].vlines(x = 16482.9, ymin = 0.75, ymax=0.85, color="purple") # PI 16482.9
    setxTicks(ax[1])
    plt.show()

    # KI lines plot
    fig, ax = plt.subplots()
    ax.set_title("KI lines")
    ax.plot(bdata["Wavelength"], bdata["Flux"], color="black", label="")
    ax.plot(apogdata["Wavelength"], apogdata["Flux"]/normk1_1, color="red")
    ax.set_xlim(*limk1)
    ax.set_ylim(0.69,1.01)
    ax.vlines(x = [15163.067, 15168,376], ymin = 0.75, ymax=0.85, color="purple") # KI 15163.067 and KI 15168.376
    setxTicks(ax)
    plt.show()
    
    # SI lines plots
    fig, ax = plt.subplots(2)
    ax[0].set_title("SI lines")
    ax[0].plot(bdata["Wavelength"], bdata["Flux"], color = "black")
    ax[0].plot(apogdata["Wavelength"], apogdata["Flux"]/norms1_1, color="red")
    ax[0].set_xlim(*lims1)
    ax[0].set_ylim(0.69,1.01)
    ax[0].vlines(x = 15469.8036, ymin = 0.7, ymax=0.8, color="purple") # SI 15469.8036
    setxTicks(ax[0])
    ax[1].plot(bdata["Wavelength"], bdata["Flux"], color = "black")
    ax[1].plot(apogdata["Wavelength"], apogdata["Flux"]/norms1_2, color="red")
    ax[1].set_xlim(*lims1_2)
    ax[1].set_ylim(0.69,1.01)
    ax[1].set_xlabel("Wavelength (A)")
    ax[1].vlines(x = [15475.604, 15478.469], ymin = 0.75, ymax=0.85, color="purple") # SI 15475.604 and SI 15478.469
    setxTicks(ax[1])
    plt.show()
```
